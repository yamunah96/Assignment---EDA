```python
# checking m value for indiviual movies
qmovies=df3.copy().loc[df3["vote_count"]>=m]
print(qmovies.shape)
print(qmovies)
```

```python
# x= df3 data
def weightRating(x,m=m,c=C):
  v=x["vote_count"]
  R=x["vote_average"]
  return (v/(v+m)*R)+(m/(m+v)*c)
qmovies["score"]=qmovies.apply(weightRating,axis=1)
qmovies.head(1)
```

```python
qmovies=qmovies.sort_values("score",ascending=False)
# qmovies.head(10)
```

```python
def getlist(x):
  if isinstance(x,list):
    names=[i['name'] for i in x]
    return names
  return []

# x=features
features=["cast","keywords","crew","genres"]
for i in features:
  df3[i]=df3[i].apply(getlist)




```

```python
def cleanData(x):
  if isinstance(x,list):
    return [str.lower(i.replace(" ",""))for i in x]
  else:
    if isinstance(x,str):
      return [str.lower(x.replace(" ",""))]
    else:
      return ""
    
# x=features
features=["cast","keywords","crew","genres"]
for i in features:
  df3[i]=df3[i].apply(cleanData)
```

```python
def join(x):
  return " ".join(x["keywords"])+" "+ " ".join(x["cast"]) +" "+ " ".join(x["genres"])+ " "+ " ".join(str(x["directors"]))
df3["soup"]=df3.apply(join,axis=1)
```

```python
# creating vector to soup
count= CountVectorizer(stop_words="english")
countMatrix=count.fit_transform(df3["soup"])
countMatrix[0]
print(count.get_feature_names())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim= cosine_similarity(countMatrix)
cosine_sim
```

```python
df3= df3.reset_index()
indices= pd.Series(df3.index,index=df3["title_x"])
df3.head()
```

```python
def getRecommendations(title,cosine_sim):
  idx= indices[title]
  sim_scores= list(enumerate(cosine_sim[idx]))
  sim_scores_sorted= sorted(sim_scores, key=lambda x:x[1], reverse=True)
  print(sim_scores_sorted)
  movielist= sim_scores_sorted[1:11]
  movieIndices= [i[0] for i in movielist]
  return df3["title_x"].iloc[movieIndices]

```

```python
getRecommendations("Harry Potter and the Order of the Phoenix",cosine_sim)
```
